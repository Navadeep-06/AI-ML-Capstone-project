```python
import re

def has_pii(text):
    """
    Check if the input contains an email address
    or a phone number.
    Returns True if PII is detected, otherwise False.
    """

    email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b"
    phone_pattern = r"\b(?:\+91[- ]?)?[6-9]\d{9}\b"

    if re.search(email_pattern, text):
        return True

    if re.search(phone_pattern, text):
        return True

    return False


def validate_prompt(prompt):
    """
    Validate the user prompt before sending it to the LLM.
    """

    if has_pii(prompt):
        return False, "PII Detected"

    return True, "Prompt is Safe"
```

```python
import re

def has_pii(text):
    email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b"
    phone_pattern = r"\b(?:\+91[- ]?)?[6-9]\d{9}\b"

    return bool(
        re.search(email_pattern, text) or
        re.search(phone_pattern, text)
    )
```

```python
print(has_pii("My email is test@gmail.com"))
# True

print(has_pii("Prediction: 1 Confidence: 0.56"))
# False
```
